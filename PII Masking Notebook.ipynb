```python
import os
import shutil

# 1. Setup paths based on your verified input
base_path = "/kaggle/input/datasets/ikramahelahi/internship"
working_dir = "/kaggle/working"

os.chdir(working_dir)

# 2. Hard cleanup of the working directory
!rm -rf ner_training.py ner_utils.py evaluate_checkpoint.py processed_data

# 3. Copy the Python modules
scripts = ['ner_training.py', 'ner_utils.py', 'evaluate_checkpoint.py']
for script in scripts:
    src = os.path.join(base_path, script)
    if os.path.exists(src):
        shutil.copy(src, os.path.join(working_dir, script))
        print(f"✅ Copied {script}")
    else:
        print(f"❌ {script} not found at {src}")

# 4. Setup the data folder
os.makedirs('processed_data', exist_ok=True)
for json_file in ['train.json', 'val.json', 'test.json']:
    src = os.path.join(base_path, json_file)
    if os.path.exists(src):
        # Using force link to avoid "File Exists" errors
        !ln -sf {src} ./processed_data/{json_file}
        print(f"🔗 Linked {json_file}")
    else:
        print(f"❌ {json_file} not found at {src}")

# 5. Quick verification
if os.path.exists("processed_data/train.json"):
    print("\n💎 Everything is ready. Launching training...")
    !python ner_training.py --epochs 5
```
